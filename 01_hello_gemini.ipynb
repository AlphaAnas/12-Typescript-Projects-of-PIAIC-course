```python
from warnings import catch_warnings
from IPython import get_ipython
from IPython.display import display
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory
from google.colab import userdata
from langchain_google_genai import GoogleGenerativeAI

llm = GoogleGenerativeAI( # Use GoogleGenerativeAI instead of GoogleGeminiFlash
    api_key = userdata.get('GOOGLE_API_KEY'),
    model='gemini-2.0-flash-exp',
    temperature=0.7  # Adjust for creativity
)


# Create a memory object to store the conversation context
memory = ConversationBufferMemory()


prompt_template = PromptTemplate(
    input_variables=["history", "input"],
     template=(
        "The following is a conversation with a helpful assistant:\n\n"
        "{history}\nHuman: {input}\nAssistant:"
    )
)

# Create the LLM chain
chain = ConversationChain(llm=llm, memory=memory, prompt=prompt_template)


while True:
    question = input("You: ")
    if question.lower() == "exit":
        print("Ending chat. Goodbye!")
        break
    response = chain.run(input=question)
    print("Assistant:", response)

```
